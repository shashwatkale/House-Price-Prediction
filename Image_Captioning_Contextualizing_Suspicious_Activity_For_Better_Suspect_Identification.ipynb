```python
from transformers import pipeline
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import time
import datetime
import pytz
from transformers import pipeline
image_to_text = pipeline("image-to-text", model="Salesforce/blip-image-captioning-large")
```

```python
indian_timezone = pytz.timezone('Asia/Kolkata')

current_time = datetime.datetime.now(indian_timezone)

current_time.strftime('%Y-%m-%d %H:%M:%S')
```

```python
def detect():
  word = ['run','running ','gun','fighting','fight','knif','mask','wepone','police','fear','punch','dance','running',"gang",'looking']

  cap = cv2.VideoCapture("/content/The Gentleman Gangster Scene.mp4")


  if not cap.isOpened():
      print("Error: Could not open video file.")
      exit()


  import os
  output_directory = 'output_frames'
  os.makedirs(output_directory, exist_ok=True)


  frame_interval = 30

  frame_count = 0

  # Read and save video frames
  while True:
      ret, frame = cap.read()

      if not ret:
          break

      if frame_count % frame_interval == 0:

          frame_filename = f"{output_directory}/frame_{frame_count:04d}.jpg"
          image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
          cv2.imwrite(frame_filename, image)


          ts = current_time.strftime('%Y-%m-%d %H:%M:%S')
          result = image_to_text(frame_filename)
          print(f"Saved frame {frame_count}")
          print(f"[{ts}] {result[0].get('generated_text')}")

          if any(x in word for x in result[0].get('generated_text').split()):
            print("[STATUS] Suspicious")
          else:
            print("[STATUS] Normal")

      frame_count += 1


  cap.release()
  cv2.destroyAllWindows()

  print(f"Saved {frame_count // frame_interval} frames to {output_directory}")
```

```python
detect()
```
